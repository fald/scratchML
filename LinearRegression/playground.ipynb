```python
from LinearRegression import LinearRegression
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
```

```python
X, y = datasets.make_regression(n_samples=1000, n_features=1, noise=20, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], y, color="b", marker="o", s=30)
plt.show()
```

```python
linReg = LinearRegression()
linReg.fit(X_train, y_train)
```

```python

predictions = linReg.predict(X_test)
error = linReg.MSE(X_test, y_test) # I used predict in my MSE method, so...yeah.
```

```python
fig = plt.figure(figsize=(8, 6))
# plt.scatter(X_train[:, 0], y_train, color="b", marker="o", s=30)
plt.scatter(X_test[:, 0], y_test, s=30, color="r", marker="x")
plt.plot(X_test[:, 0], predictions, color="black", linewidth=1, label="Predictions")
# The LR needed to be increased by a factor of 10, though iterations from 1-2k didn't seem to change much
# Looking at history, 600 iters before it petered out with LR = 1e-2
plt.show()
print(predictions)
```
